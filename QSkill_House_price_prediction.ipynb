Now we see in the price column of the data set , firstly we describe the dataset of it and then assign the index values of each price using reset_index()


```python
data["Price"].describe().reset_index()["Price"]
```

```python
stats = data["Price"].describe().reset_index()
stats["Price"] = round(stats["Price"],2) # rounding off to 2 decimal places
```

```python
# for na values we will use :
data.dropna(inplace=True)

# for duplicated values
data.drop_duplicates(inplace=True)
```

```python
data.groupby("condition of the house")["Price"].mean().sort_values(ascending=True).plot(kind = 'bar')
plt.xlabel("condition of the house")
plt.ylabel("mean value")
plt.title("conditon of the house Vs the price")
plt.show()
```
